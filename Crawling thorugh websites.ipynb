```python

from bs4 import BeautifulSoup
import ssl
import urllib.parse , urllib.error , urllib.request


#Write an SSL certificate to capture Https (Secure Netwrok Connection Required) sites as well
ctx = ssl.create_default_context()
#Not secure to set it to false
ctx.check_hostname = False 
ctx.verify_mode = ssl.CERT_NONE

#http://py4e-data.dr-chuck.net/known_by_Fikret.html t.contents[0]
x = input("Enter the position: ")
y = int(x)
w = input("Enter how many times")
z = int(w)

def openUrlFromStr(url):
    htmlFile = urllib.request.urlopen(url, context = ctx).read()
    #The soup object arranges the HTML data into an organized way
    soupParserF = BeautifulSoup(htmlFile, 'html.parser')
    #Returns a Soup Object
    return soupParserF


    
    
times = 0
lst  = list()
  
while True:
    
    if times == 0 :
        #First time
        inp = input("Enter url for the first time: ")
        spo = openUrlFromStr(inp)
        #A soup Object Retunred
        print("First spo ", spo.title)
        tags = spo('a')
        val = tags[y-1]
        nl = val.get('href', None)
        lst.append(str(nl))
        print("List",lst[0])
        times = times + 1
        print("In the first time loop", times , val , nl )
        continue
    
    elif times == z :
        #Check the last entered entry in X list
        openStr = lst.pop()
        #Split
        hyphenSplit = openStr.split('_')
        val = hyphenSplit[2].split('.')[0]
        print(val)
        break
    else:
        vi = lst[-1]
        print("Vai", vi)
        cso = openUrlFromStr(vi)
        tags = cso('a')
        val = tags[y-1]
        nl = val.get('href', None)
        lst.append(nl)
        times = times + 1
        
     
     
```
